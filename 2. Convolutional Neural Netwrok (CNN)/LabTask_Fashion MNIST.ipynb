```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.utils import to_categorical
```

```python
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#preprocess the data
X_train=X_train.reshape((X_train.shape[0],28*28)).astype('float32')/255
X_test=X_test.reshape((X_test.shape[0],28*28)).astype('float32')/255
```

```python
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=Sequential([
    Dense(120,activation='relu', input_shape=(28*28,)),
    Dense(60,activation='relu',input_shape=(28*28,)),
    Dense(40,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
test_loss,test_acc=model.evaluate(X_test,y_test)
print("Test Accuracy: ", test_acc)
```

```python
# Define class names for Fashion MNIST dataset
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Plot the images with their class labels
plt.figure(figsize=(20, 15))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i].reshape(28, 28), cmap=plt.cm.binary)  # Reshape the image to 28x28
    plt.xlabel(class_names[np.argmax(y_train[i])])  # Use argmax to find the index of the class label
plt.show()
```
